```python
import pandas as pd
# pandas is a powerful library for data manipulation and analysis
# load the dataset (tab separated values)
df = pd.read_csv("../datasets/en.openfoodfacts.org.products.tsv",
                 sep="\t", low_memory=False # to avoid dtype warnings
                    )
df.info()
```

```python
# Top 10 brands by product count
top_brands = df['brands'].value_counts().head(10)
display(top_brands)
```

```python
import matplotlib.pyplot as plt

# Bar chart of the top-10 brands you already computed
top_brands.plot.bar(figsize=(8,4))
plt.title("Top 10 Brands by Product Count in Open Food Facts")
plt.ylabel("Number of Products")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
# Save the top brands to a CSV file
```
